```python

def add_zoom_inset(ax, image, zoom_region, zoom_factor=2, inset_size=0.5, loc=1):
    x1, y1, x2, y2 = zoom_region
    
    # Add a rectangle to indicate the zoom region
    rect = patches.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=2, edgecolor='blue', facecolor='none')
    ax.add_patch(rect)
    
    # Create the zoomed inset
    zoomed_inset = image[y1:y2, x1:x2]
    ax_inset = ax.inset_axes([0.6, 0.6, inset_size, inset_size])  # Adjusted size and position for top-right corner
    ax_inset.imshow(cv2.cvtColor(cv2.resize(zoomed_inset, None, fx=zoom_factor, fy=zoom_factor, interpolation=cv2.INTER_NEAREST), cv2.COLOR_BGR2RGB))
    ax_inset.set_xticks([])
    ax_inset.set_yticks([])
    # ax_inset.set_title("Zoomed-in")


image1 = cv2.imread('output_v2_frame/AMVTG_004_original.png', cv2.IMREAD_COLOR)
image2 = cv2.imread('output_v2_frame/AMVTG_004_bicubic_x4.png', cv2.IMREAD_COLOR)
image3 = cv2.imread('output_v2_frame/AMVTG_004_srcnn_x4.png', cv2.IMREAD_COLOR)

# image1 = cv2.imread('output_v2_frame/LAND9_007_original.png', cv2.IMREAD_COLOR)
# image2 = cv2.imread('output_v2_frame/LAND9_007_bicubic_x4.png', cv2.IMREAD_COLOR)
# image3 = cv2.imread('output_v2_frame/LAND9_007_srcnn_x4.png', cv2.IMREAD_COLOR)


# Define the zoom region (top-left corner for demonstration)
zoom_region = (200, 200, 400, 400)

fig1, ax1 = plt.subplots(figsize=(6, 6))
ax1.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
ax1.set_title("Original")
ax1.axis('off')
add_zoom_inset(ax1, image1, zoom_region)
plt.tight_layout()
plt.savefig("original_image.png")
plt.show()

# Create the second plot for the bicubic image
fig2, ax2 = plt.subplots(figsize=(6, 6))
ax2.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
ax2.set_title("Bicubic")
ax2.axis('off')
add_zoom_inset(ax2, image2, zoom_region)
plt.tight_layout()
plt.savefig("bicubic_image.png")
plt.show()

# Create the third plot for the upsampled image
fig3, ax3 = plt.subplots(figsize=(6, 6))
ax3.imshow(cv2.cvtColor(image3, cv2.COLOR_BGR2RGB))
ax3.set_title("Upsampled Image")
ax3.axis('off')
add_zoom_inset(ax3, image3, zoom_region)
plt.tight_layout()
plt.savefig("upsampled_image.png")
plt.show()
```
